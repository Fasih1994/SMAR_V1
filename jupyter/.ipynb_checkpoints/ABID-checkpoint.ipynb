```python
from langchain import OpenAI
from langchain.agents import create_pandas_dataframe_agent
import pandas as pd
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
adib = pd.read_csv('~/ADIB/data/ADIB.csv',encoding='latin-1')
adib
```

```python
fab = pd.read_csv('~/ADIB/data/FAB.csv',encoding='latin-1')
fab
```

```python
adib['name']='ADIB'
adib.head()
```

```python
fab['name']='FAB'
fab.head()
```

```python
df = pd.concat([adib,fab])
df.head()
```

```python
df.tail()
```

```python
agent = create_pandas_dataframe_agent(
    OpenAI(temperature=0.6), 
    df=df, 
    verbose=True,
    max_iterations=5
)
```

```python
PROMPT = """You are a chatbot for Abu Dhabi Islamic Bank named ADIB.ai
Example Query: "Please provide the investments in the year 2021 and the investment growth."

You have balance-sheet data for different banks, identified by the `name` column. 
Answer for ADIB bank only unless instructed otherwise for specific banks.

Example Answers:
1. "The investments in the year 2021 amount to 13,691 Million AED."
2. "The investment growth is 10%."

All values are represented in Million AED, and percentages and ratios are properly formatted.
Here's the query: 
"""
def ask_agent(agent, query):
    """
    Query an agent and return the response as a string.

    Args:
        agent: The agent to query.
        query: The query to ask the agent.

    Returns:
        The response from the agent as a string.
    """
    # Prepare the prompt with query guidelines and formatting
    prompt = ( PROMPT + query)

    # Run the prompt through the agent and capture the response.
    with get_openai_callback() as cb:
        response = agent.run(prompt)
        total_tokens = cb.total_tokens
        prompt_token = cb.prompt_tokens
        completion_token = cb.completion_tokens
        total_cost = cb.total_cost
    print("total token used:", total_tokens, 'costing:', total_cost)

    # Return the response converted to a string.
    return str(response)
```

```python
question = "What was the customer financing in 2018?"
question2 = "What was the other assets 2020?"
question3 = "What was the investments in 2021?"
question4 = "What was the difference between investments of 2021 and 2019?"
```

```python
response = ask_agent(agent=agent, query=question)
print(response)
```
