# Different Section

```python
from langchain.document_loaders.csv_loader import CSVLoader


csv_loader = CSVLoader(file_path='/home/fasih/ADIB/data/ADIB.csv')
```

```python
from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import PyPDFDirectoryLoader

BASE_DIR = "/home/fasih/ADIB/pdfs/"
pdf_loader = PyPDFDirectoryLoader(BASE_DIR)
```

```python
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma

embeddings = OpenAIEmbeddings()
persist_directory = '/home/fasih/ADIB/chroma'
```

```python
csv_data = csv_loader.load()
pdf_data = pdf_loader.load()
docs = csv_data+pdf_data
```

```python
from langchain.text_splitter import CharacterTextSplitter

c_spliter = CharacterTextSplitter(
    chunk_size=1500,chunk_overlap=150,
    separator='\n'
)
```

```python
splits = c_spliter.split_documents(docs)
```

```python
splits[:5]
```

```python
vectoDB = Chroma.from_documents(
    documents=splits,
    embedding=embeddings,
    persist_directory=persist_directory
)
```

```python
vectoDB._collection.count()
```

```python
question = 'What is the cash and balance with central bank for 2021 for ADIB'
vectoDB.similarity_search(question,k=2)
```

```python
question = 'What is the cash and balance with central bank for 2021 for ADIB'
vectoDB.max_marginal_relevance_search(query=question, k=1, fetch_k=2)
```

```python
from langchain.llms import OpenAI
from langchain.retrievers.self_query.base import SelfQueryRetriever
from langchain.chains.query_constructor.base import AttributeInfo
```

```python
pdf_data[0]
```

```python
metadata_field_info = [
    AttributeInfo(
        name="source",
        description="Name of the bank in Question",
        type="string",
    ),
    AttributeInfo(
        name="source",
        description="The pdf document the chunk is from",
        type="string",
    ),
]
```

```python
document_content_description = "Bank's balance sheet and Regulations"
llm = OpenAI(temperature=0)
retriever = SelfQueryRetriever.from_llm(
    llm,
    vectoDB,
    document_content_description,
    metadata_field_info,
    verbose=True
)
```

```python
retriever.get_relevant_documents(question)
```

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
```

```python
llm = OpenAI(temperature=0)
compressor = LLMChainExtractor.from_llm(llm)
```

```python
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor,
    base_retriever=vectoDB.as_retriever(search_type='mmr')
)
```

```python
def pretty_print_docs(docs):
    print(f"\n{'-' * 100}\n".join([f"Document {i+1}:\n\n" + d.page_content for i, d in enumerate(docs)]))
```

```python
question = 'What is the cash and balance with central bank for 2021 for ADIB bank'
compressed_docs = compression_retriever.get_relevant_documents(question)
pretty_print_docs(compressed_docs)
```

```python
from langchain.prompts import PromptTemplate
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Use three sentences maximum. Keep the answer as concise as possible.{context}
Question: {question}
Helpful Answer:"""
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm,
    retriever=vectoDB.as_retriever(),
    return_source_documents=True,
    chain_type_kwargs={"prompt": QA_CHAIN_PROMPT},
    
)
```

```python
from langchain.callbacks import get_openai_callback
with get_openai_callback() as cb:
    question = "Can you tell me what RoRR means?"
    result = qa_chain({"query": question})
    print(cb)
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings.openai import OpenAIEmbeddings


embeddings = OpenAIEmbeddings()
persist_directory = '/home/fasih/ADIB/chroma'
```

```python
vectordb = Chroma(persist_directory=persist_directory, embedding_function=embeddings)
vectordb._collection.count()
```

```python
from langchain.chat_models import ChatOpenAI
llm_name = 'gpt-3.5-turbo'
llm = ChatOpenAI(model_name=llm_name, temperature=0)
```

```python
llm.predict('Hello world')
```

```python
# Build prompt
from langchain.prompts import PromptTemplate
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Keep the answer as concise as possible. And remember your name is ADIB.ai. If answer is in points give complete points. Always answer in the question's language. 
{context}
Question: {question}
Helpful Answer:"""

QA_CHAIN_PROMPT = PromptTemplate(input_variables=["context", "question"],template=template,)
```

```python
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_chain_type(llm,
                                       retriever=vectordb.as_retriever(),
                                       return_source_documents=False,
                                       chain_type_kwargs={"prompt": QA_CHAIN_PROMPT})
```

```python
question = "What were the investments for 2021"
result = qa_chain({"query": question})
result["result"]
```

```python
question = "who are you"
result = qa_chain({"query": question})
result["result"]
```

```python
result
```

```python
from langchain.memory import ConversationBufferWindowMemory
memory = ConversationBufferWindowMemory(
    memory_key="chat_history",
    k=5,
    return_messages=True
)
```

```python
from langchain.chains import ConversationalRetrievalChain

qa = ConversationalRetrievalChain.from_llm(
    llm,
    retriever=vectordb.as_retriever(search_type='mmr'),
    memory=memory,
)
```

```python
from langchain.callbacks import get_openai_callback
def ask_agent(query:str=None):
    with get_openai_callback() as cb:
        result = qa({"question": query})
        print(cb)
    return result
```

```python
ask_agent('Who are you?')['answer']
```

```python
ask_agent("Why the guidance note must be read?",)['answer']
```

```python
ask_agent("what will happen if i don't read them?")['answer']
```
