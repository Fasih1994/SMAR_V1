```python
from langchain.vectorstores import Chroma
from langchain.embeddings.openai import OpenAIEmbeddings


embeddings = OpenAIEmbeddings()
persist_directory = '/home/fasih/ADIB/chroma3'
```

```python
vectordb = Chroma(persist_directory=persist_directory, embedding_function=embeddings)
vectordb._collection.count()
```

```python
from langchain.chat_models import ChatOpenAI
llm_name = 'gpt-3.5-turbo'
llm = ChatOpenAI(model_name=llm_name, temperature=0)
```

```python
llm.predict('Hello world')
```

```python
# Build prompt
from langchain.prompts import PromptTemplate
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Keep the answer as concise as possible. And remember your name is ADIB.ai. If answer is in points give complete points. Always answer in the question's language. 
{context}
Question: {question}
Helpful Answer:"""

QA_CHAIN_PROMPT = PromptTemplate(input_variables=["context", "question"],template=template,)
```

```python
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_chain_type(llm,
                                       retriever=vectordb.as_retriever(),
                                       return_source_documents=True,
                                       chain_type_kwargs={"prompt": QA_CHAIN_PROMPT})
```

```python
question = "by what percent our non funded income grew?"
result = qa_chain({"query": question})
result["result"]
```

```python
question = "what is our bank balance and cashin  Abridged Income Statement ?"
result = qa_chain({"query": question})
result["result"]
```

```python
from langchain.memory import ConversationBufferWindowMemory
memory = ConversationBufferWindowMemory(
    memory_key="chat_history",
    k=5,
    return_messages=True
)
```

```python
from langchain.chains import ConversationalRetrievalChain

qa = ConversationalRetrievalChain.from_llm(
    llm,
    retriever=vectordb.as_retriever(search_type='mmr'),
    memory=memory,
)
```

```python
from langchain.callbacks import get_openai_callback
def ask_agent(query:str=None):
    with get_openai_callback() as cb:
        result = qa({"question": query})
        print(cb)
    return result
```

```python
ask_agent('Can you tell me what RoRR means?')['answer']
```

```python
ask_agent("For MPM Properties LLC what are the total fees and commisions for 2022?",)['answer']
```

```python
ask_agent("What is year 2022 growth in net profit?")['answer']
```
